```python
from collections import defaultdict
```

```python
Cluster_doc=defaultdict(list)
```

```python
with open('Tweets_active_cluster_ICF_CWW_ALPHA0.002_BETA0.0004_LAMDA6e-06.txt') as input:
    line=input.readline()
    while line:
        obj = json.loads(line)
        for key , value in obj.items():
            Cluster_doc[int(key)].extend(value['0'])
           
            break
        line = input.readline()
        
```

```python
Cluster_doc
```

```python
Cluster_doc1=Cluster_doc.copy()
Cluster_doc1
```

```python
for edge in m_sim_cl:
    first=min(edge[1],edge[2])
    second=max(edge[1],edge[2])
    if second in Cluster_doc and first in Cluster_doc:
        Cluster_doc[first].extend(Cluster_doc[second])
        del Cluster_doc[second]
```

```python
len(Cluster_doc)
```

```python
len(Cluster_doc1)
```

```python
Cluster_doc
```

```python
with open("cluster_bench_w2v5.json", "w") as f:

        for key,value in Cluster_doc.items():
            for doc in value:
                st = ""+str(doc)+" "+str(key)+" \n"
                f.write(st)
```
